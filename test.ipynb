```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Create Arima Dataframe


```python


n = 300
# create date range from 1-1-2025 to 300 date
dates = pd.date_range('2025-01-01', periods=n, freq="D")

# create an array of random 300 values
errors = np.random.normal(0, 1, n)

# Generate ARIMA(2,1,1) process manually
# y_t = 0.6*y_{t-1} + 0.3*y_{t-2} + 0.4*e_{t-1} + e_t + trend

# create an array of zeros (float)
values = np.zeros(n)

# To create an array
trend = 0.02 * np.arange(n)

for t in range(2, n):
    # AR(2) component + MA(1) component
    ar = 0.6 * values[t-1] + 0.3 * values[t-2]
    ma = 0.4 * errors[t-1]
    values[t] = ar + ma + errors[t] + trend[t]

df = pd.DataFrame({
    'date': dates,
    'temperature': values 
})

df.set_index('date', inplace=True)
```

```python

print("="*60)
print("OUR SYNTHETIC TEMPERATURE DATA")
print("="*60)
print(f"Shape: {df.shape}")
print(f"Date range: {df.index[0]} to {df.index[-1]}")
print("\nFirst 10 days:")
print(df.head(10))
print("\nLast 10 days:")
print(df.tail(10))
```
