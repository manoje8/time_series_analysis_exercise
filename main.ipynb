```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import month_plot, quarter_plot
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
path="/home/mano/Manoj/Learning/time_series/data/bitcoin_price.csv"
```

```python
df = pd.read_csv(path)
```

```python
df.head()
```

*Convert Date object into date time*

```python
df.info()
```

```python
df["Date"] = pd.to_datetime(df["Date"], format="%Y-%m-%d")
df.set_index("Date", inplace=True)
```

```python
df.index
```

```python
df.loc['2021-11']
```

```python
# Load data and setting index
df1 = pd.read_csv(path, index_col="Date", parse_dates=True)
df1.index
```

```python
# resamplying to monthly frequency
df.resample("ME").mean()
```

```python
#7 day rolling average
df["7_day_rolling"] = df["Close"].rolling(window=7).mean()
```

```python
df[["Close", "7_day_rolling"]].loc['2021'].plot()
plt.show()
```

```python
df.resample("ME").mean()['Close'].idxmax()
```

```python
# Calculate Daily return
df['daily_returns'] = df["Close"].pct_change() * 100
```

```python
# day with more than 10% change in closing
df[abs(df['daily_returns']) > 10]
```

# Data visualization

```python
# Daily closing price
df["Close"].plot(title="Daily closing price")
plt.show()
```

```python
# Plot the yearly volume
df.resample('YE').sum()["Volume"].plot(title="Yearly Volume")
plt.show()
```

```python
# plotting closing date and 30 day rolling volume
df['30_day_rolling_vol'] = df["Volume"].rolling(window=30).mean()
df['30_day_rolling_vol'].plot(legend=True)
ax = df['Close'].plot(secondary_y=True, legend=True)
ax.set_ylabel("Volume")
plt.show()
```

```python
# correlation
df[["Close", "30_day_rolling_vol"]].corr()
```

# Data Manipulation


```python
# Identify missing value
df.isnull().sum()
```

```python
# Fill missing values
df["30_day_rolling_vol"].bfill(inplace=True)

# Interpolate missing value
df['30_day_rolling_vol'].interpolate()
```

```python
df.isnull().sum()
```

```python
df['year'] = df.index.year
df['month'] = df.index.month
df['day'] = df.index.day
df['weekday'] = df.index.day_name()
df['weekday_numeric'] = df.index.weekday
df['is_week_end'] = df.index.weekday > 4

df.head()
```

```python
# Feature Engineering
df['closed_lag1'] = df['Close'].shift(1)
df['closed_lag2'] = df['Close'].shift(2)
```

```python
df.head()
```

```python
#Plotting
month_plot(df_choco['revenue'], ylabel='Revenue')
plt.show()
```

```python
# Seasonal decomposition for bit coin data
decomposition = seasonal_decompose(df['Close'], model="multiplicative", period=365)
fig = decomposition.plot()
fig.set_size_inches(10, 8)
plt.show()
```

```python
# Seasonal decomposition for choco data
decomposition_choco = seasonal_decompose(df_choco['revenue'], model="multiplicative", period=5)
fig = decomposition_choco.plot()
fig.set_size_inches(10, 8)
plt.show()
```

Seasonality:
* 24 hrs for modeling
* 7 or 365 for daily, but 7 is preferred for modeling
* 52 for weekly
* 12 for monthly
* 4 for quaterly
* 5 for weekdays

# Auto correlation

```python
fig, ax = plt.subplots(figsize=(10,6))
plot_acf(df['Close'], lags=100, ax=ax)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,6))
plot_acf(df_choco['revenue'], lags=30, ax=ax)
plt.show()
```

# Auto partial correlation

```python
fig, ax = plt.subplots(figsize=(10,6))
plot_pacf(df['Close'], lags=100, ax=ax)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,6))
plot_pacf(df_choco['revenue'], lags=30, ax=ax)
plt.show()
```
